```python
from selenium import webdriver
```

```python
url = 'https://eungdapso.seoul.go.kr/Shr/Shr01/Shr01_lis.jsp'

driver = webdriver.Chrome('chromedriver')
driver.get(url)
```

```python
# id 속성으로 찾기
```

```python
selected_id = driver.find_element_by_id('container_sub')
print(selected_id)         # 세션 아이디
```

```python
print(selected_id.tag_name)        # 세션 아이디가 가지고 있는 태그 이름
```

```python
print(selected_id.text)            # 값
```

```python
# 태그명으로 찾기
selected_tag_name = driver.find_element_by_tag_name('a')      # 첫번째 a tag만 찾아줌
print(selected_tag_name)
print(selected_tag_name.tag_name)
print(selected_tag_name.text)
```

```python
selected_tag_name = driver.find_elements_by_tag_name('a')
print(selected_tag_name) # 같은 이름을 가진 태그는 중복가능하므로 리스트로 return
```

```python
# 인덱싱 또는 반복문으로 추출
for tag_a in selected_tag_name:
    print(tag_a.text)
```

```python
# name 속성으로 찾기

# 보통 name 속성은 form이나 inpute tag에 같이 많이 사용
selected_name = driver.find_element_by_name('StartDate_out')
print(selected_name)
print(selected_name.tag_name)
print(selected_name.text)
```

```python
selected_name = driver.find_elements_by_name('StartDate_out')
print(selected_name)
```

```python
# link 내용으로 찾기
selected_link = driver.find_element_by_link_text('')
print(selected_link)
print(selected_link.tag_name)
print(selected_link.text)
```

```python
selected_link = driver.find_elements_by_link_text('')
print(selected_link)
```

```python
for link in selected_link:
    print(link.text)
```

```python
# 데이터 전송
from selenium import webdriver
from selenium.webdriver.common.keys import Keys


url = 'https://www.naver.com'

driver = webdriver.Chrome('chromedriver')
driver.get(url)

select_query = driver.find_element_by_css_selector('input#query')
select_query.send_keys('취업')
select_query.send_keys(Keys.ENTER)
```
